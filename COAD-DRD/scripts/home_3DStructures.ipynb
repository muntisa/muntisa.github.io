### Create interactive 3D molecular complex from Vina results

```python
import nglview as nv
view = nv.NGLWidget()
```

```python
view.add_component(nv.FileStructure("extras/4gls.pdb.prot.pdbqt"))
view.add_component(nv.FileStructure("extras/out.4gls_venetoclax-49846579.pdbqt"))
```

```python
#nv.show_structure_file(nv.datafiles.PDB)
view._remote_call("setSize", target="Widget", args=["800px", "600px"])
view.center()
view
```

```python
nv.write_html('../extras/4gls_venetoclax-49846579_docking.html',[view] )
```
